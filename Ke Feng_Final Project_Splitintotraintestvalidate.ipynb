```python
#Import Libraries 
#Basic 
import pandas as pd
import numpy as np

#Visualization
import matplotlib.pyplot as plt

#processing
from sklearn.model_selection import train_test_split
```

```python
###Load the data
#The Crime Sample is pre-cleaned data 
crimesamp=pd.read_csv('crimesamp.csv', index_col=False, delimiter=',')
crimesamp.head()
```

```python
#####Data Cleaning and Manipulation
#Check missing values 
crimesamp.dropna()
```

```python
#Transform Data into Numeric
crimesamp['Description'] = pd.factorize(crimesamp["Description"])[0]
crimesamp['Location Description'] = pd.factorize(crimesamp["Location Description"])[0]
crimesamp['Arrest'] = pd.factorize(crimesamp["Arrest"])[0] 
crimesamp['Domestic'] = pd.factorize(crimesamp["Domestic"])[0] 
crimesamp.head()
```

```python
#Categorize Target based on Severity, Create a binary problem within the target class
severecrime=["ARSON", "ASSAULT", "BATTERY", "CRIM SEXUAL ASSAULT", "CRIMINAL DAMAGE", "CRIMINAL TRESPASS", "HOMICIDE", "ROBBERY"]
crimesamp["Severity"]=np.where(crimesamp['Primary Type'].isin(severecrime), 1, 0)
cleaneddf = crimesamp.drop(['Primary Type'], axis=1)
cleaneddf.head()
```

```python
#Split Cleaneddf into Train, Test, and Validate 
traindata=cleaneddf.sample(13000)
valdata=cleaneddf.sample(3000)
testdata=cleaneddf.sample(4000)
```

```python
#Export Train_Test_Validate
traindata.to_csv('traindata.csv')
valdata.to_csv('valdata.csv')
testdata.to_csv('testdata.csv')
```
