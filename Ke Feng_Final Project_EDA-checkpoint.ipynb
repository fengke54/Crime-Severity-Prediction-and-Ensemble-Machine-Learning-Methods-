```python
#####IMPORT LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import folium
```

```python
#####Initial Data Loading
#Load the data
df = pd.read_csv('crime.csv', index_col=False, delimiter=',')
df.head()
```

```python
#Check the data
df.info()
```

```python
#####Data Preprocessing
#Check and Drop Missing Values 
df.isnull().sum()
df=df.dropna()
df.isnull().sum()
```

```python
#Data Sampling 
df2008=df.loc[df['Year'] == 2008]
df2009=df.loc[df['Year'] == 2009]
df2010=df.loc[df['Year'] == 2010]
df2011=df.loc[df['Year'] == 2011]
df2012=df.loc[df['Year'] == 2012]
dfsamp=df2008
dfsamp=df2008.append(df2009)
dfsamp=dfsamp.append(df2010)
dfsamp=dfsamp.append(df2011)
dfsamp=dfsamp.append(df2012)
dfsamp.head()
```

```python
#####EDA
#Primary Types of Crime
plt.figure(figsize=(14,10))
plt.title('Crime in Each Type')
plt.ylabel('Crime Type')
plt.xlabel('Number of Crimes')
dfsamp.groupby([dfsamp['Primary Type']]).size().sort_values(ascending=False).plot(kind='bar')
plt.show()
```

```python
#Transform the Date
#Convert the dates to Pandas Datetime Format
dfsamp.Date=pd.to_datetime(dfsamp.Date, format='%m/%d/%Y %I:%M:%S %p')
#Setting the index for the time
dfsamp.index = pd.DatetimeIndex(dfsamp.Date)
```

```python
#Crime Type Occurance By Year 
crimebydate = dfsamp.pivot_table('ID', aggfunc=np.size, columns='Primary Type', index=dfsamp.index.date, fill_value=0)
crimebydate.index = pd.DatetimeIndex(crimebydate.index)
plot = crimebydate.rolling(365).sum().plot(figsize=(30, 20), subplots=True, layout=(-1, 5), sharex=False, sharey=False)
```

```python
#Geographical Mapping on Arrest
plt.figure(figsize=(20,20))
sns.lmplot(x='X Coordinate', y='Y Coordinate', size=10, hue='Arrest', data=dfsamp, fit_reg=False)
plt.ylabel('Longitude', fontsize=12)
plt.xlabel('Latitude', fontsize=12)
plt.title('Arrest in Chicago')
plt.show()
```

```python
#Geographical Mapping 
MapOfChicago = folium.Map(location=[41.881832, -87.623177],
                         zoom_start=11)
dfsamp = dfsamp.dropna()
dfsamp = dfsamp.drop(columns=['Block', 'IUCR','Domestic', 'Beat', 'District', 'Ward','X Coordinate', 'Y Coordinate','Updated On', 'FBI Code'], axis = 1)
Theft = dfsamp[dfsamp["Primary Type"] == "THEFT"]
locations = dfsamp.groupby('Community Area').first()
new_locations = locations.loc[:, ['Latitude', 'Longitude', 'Location Description', 'Arrest']]
for i in range(len(new_locations)):
    lat = new_locations.iloc[i][0]
    long = new_locations.iloc[i][1]
    popup_text = """Community Index : {}<br>
                Arrest : {}<br>
                Location Description : {}<br>"""
    popup_text = popup_text.format(new_locations.index[i],
                               new_locations.iloc[i][-1],
                               new_locations.iloc[i][-2]
                               )
    folium.CircleMarker(location = [lat, long], popup= popup_text, fill = True).add_to(MapOfChicago)
MapOfChicago
```
